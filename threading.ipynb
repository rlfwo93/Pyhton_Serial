```python
def sum(low, high):
    total = 0
    for i in range(low, high):
        total += i
    print("Subthread", total, end ="\n\n")
 
t = threading.Thread(target=sum, args=(1, 100000))
t.start()
 
print("Main Thread", end ="\n\n")
```

```python
#멀티스레드 run 미구현
re = 0
ser = serial.Serial(port = "COM4", baudrate=4800, parity=serial.PARITY_NONE, stopbits=serial.STOPBITS_ONE, bytesize=serial.EIGHTBITS)
lock = threading.Lock()  
list_1=[]
need =[]
print("시작점")
def callmyname(ser):
    global list_1
    global need
    
    while ser.readable():
        print("서브스레드2")
        lock.acquire()
        res = ser.readline()
        rep = res.decode()[:len(res)-2]
        list_1.append(rep)
        print("원본 값 = {0}".format(rep))
        time.sleep(1)
        
        print(rep)
        lock.release()
        
print(__name__)
if __name__ == "__main__":
    
    print("메인스레드")
    
    t= threading.Thread(target = callmyname, args=(ser,))
    t.start() #서브스레드 시작 
     
    
    time.sleep(2)
    
    for i in range(0,len(list_1)):
        time.sleep(1) 
        lock.acquire()
        print("메인스레드 p3")
        if list_1[i][5] =="C" and list_1[i][18] =="V":
            need.append(list_1[i])
            print("필요한 값 = {0}".format(need)) 
        

        lock.release()
        
    
    
    
```
